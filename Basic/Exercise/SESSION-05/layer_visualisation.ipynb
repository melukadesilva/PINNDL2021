```python
def forward_img(model, x, layer_max_count):
    deconv_layers_list = []
    unpool_layers_list = []

    layer_count = 0

    for layer in model.features:
        if isinstance(layer, torch.nn.Conv2d):
            B, C, H, W = x.shape
            x = layer(x)
            deconv_layer = nn.ConvTranspose2d(layer.out_channels, C, layer.kernel_size, layer.stride, layer.padding)
            deconv_layer.weight = layer.weight
            assert torch.equal(deconv_layer.weight, layer.weight), "Different weights"
            deconv_layers_list.append(deconv_layer)

        if isinstance(layer, torch.nn.ReLU):
            x = layer(x)
            deconv_layers_list.append(layer)

        if isinstance(layer, torch.nn.MaxPool2d):
            x, index = layer(x)
            unpool_layers_list.append(index)
            unpool_layer = torch.nn.MaxUnpool2d(kernel_size=layer.kernel_size, stride=layer.stride,
                                                padding=layer.padding)
            deconv_layers_list.append(unpool_layer)

        layer_count += 1
        if layer_max_count == layer_count:
            break

    return x, deconv_layers_list, unpool_layers_list
```

```python
def filter_feature_maps(raw_feature_maps):
    feature_maps = raw_feature_maps[0]
    # print(feature_maps.shape)
    feature_maps_total_num = feature_maps.shape[0]

    activation_list = []
    for i in range(feature_maps_total_num):
        # For each channel get the max activation
        activation_val = torch.max(feature_maps[i, :, :])
        # print(activation_val)
        activation_list.append(activation_val.item())

    # Sort the activations along with their indices
    idx_sorted = np.argsort(activation_list)[::-1] # Reverse the list for descending order
    idx_top_k = idx_sorted[:1]
    print(idx_top_k)

    # Find the feature map index that has the highest activation value
    max_map_num = np.argmax(np.array(activation_list))
    print(max_map_num)
    # Get that feature map
    max_map = feature_maps[max_map_num, :, :]
    max_activation_val = torch.max(max_map)
    print(max_activation_val)
    max_map = torch.where(max_map == max_activation_val,
                          max_map,
                          torch.zeros(max_map.shape)
                          )
    # print(max_map)
    for i in range(feature_maps_total_num):
        if i != max_map_num:
            feature_maps[i, :, :] = 0
        else:
            feature_maps[i, :, :] = max_map
    print(feature_maps.shape)
    return feature_maps.unsqueeze_(0)
```

```python
def filter_top_k_feature_maps(raw_feature_maps, top_k):
    feature_maps = raw_feature_maps[0]
    # print(feature_maps.shape)
    feature_maps_total_num = feature_maps.shape[0]

    activation_list = []
    for i in range(feature_maps_total_num):
        # For each channel get the max activation
        activation_val = torch.max(feature_maps[i, :, :])
        # print(activation_val)
        activation_list.append(activation_val.item())

    
    # Sort the activations along with their indices
    idx_sorted = np.argsort(activation_list)[::-1] # Reverse the list for descending order
    idx_top_k = idx_sorted[:top_k]
    # print(idx_top_k)
    # store the max maps
    max_map_dict = dict()
    for idx in idx_top_k:
      # Get that feature map
      max_map = feature_maps[idx, :, :]
      max_activation_val = torch.max(max_map)
      # print(max_activation_val)
      max_map_dict[idx] = torch.where(max_map == max_activation_val,
                            max_map,
                            torch.zeros(max_map.shape)
                            )
    # print(max_map)
    for i in range(feature_maps_total_num):
        if i in idx_top_k:
          # print(i)
          feature_maps[i, :, :] = max_map_dict[i]
        else:
          feature_maps[i, :, :] = 0

    # print(feature_maps.shape)
    return feature_maps.unsqueeze_(0)
```

```python
def backward_feature_maps(y, deconv_layers_list, unpool_layers_list):
    # print(y.shape)
    for l in deconv_layers_list:
      print(l)
    for layer in reversed(deconv_layers_list):
        if isinstance(layer, nn.MaxUnpool2d):
            y = layer(y, unpool_layers_list.pop())
        else:
            # print("in",y.shape)
            y = layer(y)
            # print("out",y.shape)

    return y
```

```python
def visualize(layer_max_count, img, orig_img):
    npimg = img[0].data.numpy()
    # print(npimg.shape)
    npimg = ((npimg - npimg.min()) * 255 / (npimg.max() - npimg.min())).astype('uint8')
    npimg = np.transpose(npimg, (1, 2, 0))
    path = "/content/drive/MyDrive/Test/" + str(layer_max_count) + "th-layer.png"

    # npimg = np.uint8(npimg * (0.5) + orig_img * (0.5))
    # npimg = cv2.resize(npimg, (224, 224))
    plt.imshow(npimg)
    plt.show()
    plt.imsave(path, npimg)
```
